```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
plt.style.use('./deeplearning.mplstyle')
from lab_utils_common import dlc
from lab_coffee_utils import load_coffee_data, plt_roast, plt_prob, truncate_colormap, plt_layer, plt_network, plt_output_unit
import logging
logging.getLogger("tensorflow").setLevel(logging.ERROR)
tf.autograph.set_verbosity(0)
```

```python
X,Y = load_coffee_data()
```

```python
print(X.shape)
print(Y.shape)
```

```python
#Create a normalization layer that will be used to standardize the input data
print(f"max and min temp in pre normalized: {np.max(X[:,0]):0.2f}, {np.min(X[:,0]):0.2f}")
print(f"max and min duration in pre normalized: {np.max(X[:,1]):0.2f}, {np.min(X[:,1]):0.2f}")
norm_1 = tf.keras.layers.Normalization(axis=-1)
norm_1.adapt(X) #compute mean and variance for the data X(for transforming the data to have a mean of 0 and a standard deviation of 1.)
Xn = norm_1(X)  #normalizing the data

print(f"max and min temp in pre normalized: {np.max(Xn[:,0]):0.2f}, {np.min(Xn[:,0]):0.2f}")
print(f"max and min duration in pre normalized: {np.max(Xn[:,1]):0.2f}, {np.min(Xn[:,1]):0.2f}")
```

Tile/copy our data to increase the training set size and reduce the number of training epochs.

```python
Xt = np.tile(Xn, (1000, 1))
Yt = np.tile(Y, (1000,1))
print(Xt.shape)
print(Yt.shape)
```

Tensorflow model

```python
tf.random.set_seed(1234)
```

```python
model = Sequential(
    [
         tf.keras.Input(shape=(2,)),
         Dense(3, activation='sigmoid', name='layer_1'),
         Dense(1, activation='sigmoid', name='layer_2')
    ]
   
 )
#shape=(2,) it indicates tuple with single number for the shape of input
#shape=(2) it indicates with single number for the shape of input
```

```python
model.summary()
```

```python
L1_num_params = 2 *3 + 3  # W1 parameters  + b1 parameters(w = number of features in input, number of units in the layer)
L2_num_params = 3 *1 + 1   # W2 parameters  + b2 parameters(the bias  ð‘ size should match the number of units in the layer)
print(L1_num_params)
print(L2_num_params)
#In the first layer with 3 units, we expect W to have a size of (2,3) and  ð‘should have 3 elements.
# In the second layer with 1 unit, we expect W to have a size of (3,1) and  ð‘ should have 1 element.
```

```python
W1, b1 = model.get_layer("layer_1").get_weights()
W2, b2 = model.get_layer("layer_2").get_weights()
print(f"W1{W1.shape}: \n", W1, f"b1{b1.shape}:\n", b1)
print(W2)
print(b2)
```

1.model.compile --> statement defines a loss function and specifies a compile optimization


2.model.fit --> statement runs gradient descent and fits weights into data

```python
model.compile(
    loss = tf.keras.losses.BinaryCrossentropy(),
    optimizer = tf.keras.optimizers.Adam(learning_rate=0.01)
)
model.fit(
    Xt,Yt,epochs=10

)
```

```python
#after fitting weights has been updated
W1, b1 = model.get_layer("layer_1").get_weights()
W2, b2 = model.get_layer("layer_2").get_weights()
```

```python
print(W1)
print(W2)
```

Prediction:Once you have a trained model, you can then use it to make predictions. Recall that the output of our model is a probability. In this case, the probability of a good roast. To make a decision, one must apply the probability to a threshold. In this case, we will use 0.5

```python
X_test = np.array([
    [200,13.9],
    [200,17]             
])
X_testn = norm_1(X_test)
predictions = model.predict(X_testn)
print(f"predictions:{predictions}")
```

To convert probabilities to a decision we apply threshold

```python
y_hat = np.zeros_like(predictions)
for i in range(len(predictions)):
    if predictions[i]>= 0.5:
        y_hat[i] = 1
    else:
        y_hat[i] = 0
        
print(f"decisions = \n{y_hat}")
```

```python
#or we can achieve from this also
y_hat = (predictions>=0.5).astype(int)
print(y_hat)
```
